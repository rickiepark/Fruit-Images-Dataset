```python
import os
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

```python
dirs = ['Training/Apple Red 1', 
        'Training/Banana Lady Finger',
        'Training/Onion White', 
        'Training/Pineapple', 
        'Training/Raspberry']

label_index = 0
data = []
label = []

# save_flag = 1
for d in dirs:
    for file in os.scandir(d):
        if file.name[-3:] != 'jpg':
            continue
        img = io.imread(file.path)
        data.append(img)
        label.append(label_index)
    label_index += 1

data = np.array(data)
label = np.array(label)
```

```python
gray_data = np.dot(data, [0.2989, 0.5870, 0.1140])
```

```python
gray_data = gray_data.astype('uint8')
```

```python
plt.imshow(gray_data[1000], cmap='gray')
```

```python
with open('fruits_gray_data.npy', 'wb') as f:
    np.save(f, gray_data)

with open('fruits_data.npy', 'wb') as f:
    np.save(f, data)

with open('fruits_label.npy', 'wb') as f:
    np.save(f, label)
```
